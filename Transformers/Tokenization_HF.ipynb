```python
#THe Autotokenizer is similar to AutoModel and can grab a tokenizer class directly form the library receiving only the checkpoint name

from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
tokenizer("Using a Transformer network is simple")
#{'input_ids': [101, 7993, 170, 13809, 23763, 2443, 1110, 3014, 102], 'token_type_ids': [0, 0, 0, 0, 0, 0, 0, 0, 0], 'attention_mask': [1, 1, 1, 1, 1, 1, 1, 1, 1]}
```

```python
#Again saving a tokenizer is the same way as with the models

tokenizer.save_pretrained("directory_on_my_computer")
```

```python
#But how the Input IDs are generated? let's take a look at Encoding in order to understand the intermediate methods of the tokenizer
```

```python
#Is important to remember here the steps that happen when tokenizing a text

#raw text -> TOkens -> Special tokens -> Input Ids

#Let's see the first part raw_test -> Tokens

tokens = tokenizer.tokenize("Let's try to tokenize!")
print(tokens)
#['Let', "'", 's', 'try', 'to', 'token', '##ize', '!']
```

```python
#We can see the tokenizer performs some operations and include the ## in the subwords tokens that it has found. THis can vary from the tokenizer of one model to another. For example the albert tokenizer will add a _ in front of every token that is start of word
```

```python
#Now we need to map each token to its correspondant id. TO do so we'll need to use convert_tokens_to_ids

input_ids = tokenizer.convert_tokens_to_ids(tokens)
print(input_ids)
#[2421, 112, 188, 2222, 1106, 22559, 3708, 106]
```

```python
#The upside representation corresponds to the tokens themselves but not to the final representation so we need to add the special_tokens

final_inputs = tokenizer.prepare_for_model(input_ids)
print(final_inputs["input_ids"])
#[101, 2421, 112, 188, 2222, 1106, 22559, 3708, 106, 102]
```

```python
#In order to see the real tokenization of the transformer tokenizer you can run the .decode() method

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
inputs = tokenizer("Let's try to tokenize!")

print(tokenizer.decode(inputs["input_ids"]))
```

```python
#You can see that the sentence has 2 special tokens that are prepared when using the prepare_for_model method.
#This special tokens indicate start and end of sentence being that in this case we only gave this sentence.
# Other tokenizers use different nomeclature to split words like the one form Robert that uses html tags
```
